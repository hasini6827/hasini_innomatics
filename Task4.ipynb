**1. Mobile Recharge Validation System**

```python
def mobile_recharge(amount):
    valid_plans = [199, 299, 399, 599]
    if amount < 50:
        return False
    elif amount not in valid_plans:
        return False
    else: 
        return True
while True:
    amount = int(input("Enter recharge amount:"))
    if mobile_recharge(amount):
        print("Recharge Successful")
        break
    else:
        print("Invalid recharge amount")
                 
```

**2. Inventory Reorder Alert System**

```python
def inv(products):
    for item in products:
        if products[item] < 15:
            print(item,"Reorder Alert")
        else:
            print(item,"Stock OK")
products = {
    "Pencil" : 20,
    "Pen" : 30,
    "Eraser" : 10,
    "Sharpner" : 15
}
inv(products)
```

**3. Student Result Processing System**

```python
def checking_result(marks):
    total = 0
    for mark in marks:
        total += mark
    avg = total / len(marks)
    if avg >= 50:
        return " Pass "
    else:
        return "Fail"
marks_list = [50, 45, 34,78]
result = checking_result(marks_list)
print("Result:", result)
```

**4. Cab Fare Estimator with Retry Option**

```python
def calculate_fare(distance, peak):
    base_fare = 50
    per_km = 12
    fare = base_fare + (distance * per_km)
    if peak == "yes":
        fare += fare * 0.25
    return fare
while True:
    distance = float(input("Enter distance:"))
    peak = input("Is it peak hour?(yes/no): ").lower()
    total_fare = calculate_fare(distance, peak)
    print("Total Fare: Rupees ", total_fare)
    retry = input("Do you want to calculate again? (yes/no): ").lower()
    if retry != "yes":
        break
```

**5. Employee Attendance Eligibility Checker**

```python
def eligibility(attendance):
    present = 0
    for day in attendance:
        if day == "P":
            present += 1
    percentage = (present / len(attendance)) * 100
    if percentage >= 75:
        return "Eligible"
    else:
        return " Not Eligible"
attendance_list = ["P", "P", "A", "A", "P", "P"]
status = eligibility(attendance_list)
print("Attendance Status:", status)
```

**6. Password Strength Checker**

```python
def check_password(password):
    has_digit = False
    has_special = False
    if len(password) < 8:
        return "Weak Password"
    for ch in password:
        if ch.isdigit():
            has_digit = True
        if ch in"@$#":
            has_special = True
    if has_digit and has_special:
        return "Strong Password"
    else:
        return "Weak Password"
password = input("Enter password:")
print(check_password(password))
```
