**Problem 1: Social Media - Post Engagement Analyzer**

```python
def social_media (likes):
    total_likes = 0
    for like in likes:
        total_likes += like
    print("Total likes:", total_likes)
    if total_likes >= 1000:
        print("Viral Post")
    else:
        print("Normal Engagement")
like = [200, 450, 100, 300]
social_media(like)
```

**Problem 3: Agriculture - Rainfall Adequacy Checker**

```python
def check_rainfall(rainfall_data, required_levels):
    total = 0
    for rain in rainfall_data:
        total += rain
    average = total / len(rainfall_data)
    print("Average Rainfall:", average)
    if average >= required_levels:
        print("Rainfall Status: Adequate Rainfall")
    else:
        print("Rainfall Status: Inadequate Rainfall")
rainfall = [70, 80, 65, 73]
check_rainfall(rainfall, 70)
```

**Problem 4: Social Media - Duplicate Account Detection**

```python
def check_duplicate_accounts(usernames):
    unique_users = set(usernames)
    if len(unique_users) < len(usernames):
        print("Duplicate Account Found: Yes")
    else:
        print("Duplicate Accounts Found: No")
users = ["hasini", "akshara", "akshaya","hasini"]
check_duplicate_accounts(users)
```

**Problem 6: Agriculture- Premium Crop Price Filter**

```python
def filter_premium_crops(prices):
    premium = []
    for price in prices:
        if price > 2000:
            premium.append(price)
    print("Premium Crops:",premium)
prices = [1500, 2500, 3200, 1800]
filter_premium_crops(prices)
```

**Problem 7: System Monitoring – Application Health Checker**

```python
def check_system_health(errors):
    print("Error Count:",errors)
    if errors == 0:
        print("System Status: Healthy")
    elif errors <= 5:
        print("System Status: Minor Issues")
    else:
        print("System Status: Critical Issues")
check_system_health(7)
```

**Problem 9: E-Learning – Student Attendance Eligibility System**

```python
def check_attendance(attended_classes, total_classes):
    percentage = (attended_classes / total_classes)*100
    print("Attendance Percentage:", percentage)
    if percentage >= 75:
        print("Exam Eligibility: Eligible")
    else:
        print("Exam Eligibility: Not Eligible")
check_attendance(16, 20)
```

**Problem 10: Smart Electricity Bill Analyzer**

```python
def calculate_bill(units):
    bill = 0
    if units <= 100:
        bill = units*3
    elif units <= 200:
        bill = (100*3) + (units - 100)*5
    else:
        bill = (100*3) + (100*5) + (units -200) *7
    if bill < 500:
        usage = "Low Usage"
    elif bill <= 1500:
        usage = "Moderate Usage"
    else:
        usage = "High Usage"
    print("Total Bill :",bill)
    print("Usage Level:",usage)
calculate_bill(250)
```
