Problem Statement 1:Smart Parking Lot Management System


```python
def manage_parking(capacity, logs):
    current_count = 0
    peak_count = 0

    for entry in logs:
        if entry == "IN":
            current_count += 1
        elif entry == "OUT":
            if current_count > 0:
                current_count -= 1

        # Track peak usage
        if current_count > peak_count:
            peak_count = current_count

    print("Currently Parked Vehicles:", current_count)
    print("Peak Parking Usage:", peak_count)

    if current_count > capacity:
        print("Parking Status: Full - Capacity Exceeded")
    else:
        print("Parking Status: Available")


# Sample Run
capacity = 50
logs = ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"]
manage_parking(capacity, logs)
```

 Problem Statement 2:Online Food Delivery Time Estimator


```python
def estimate_delivery_time(distance, traffic, weather):
    base_speed = 40  # average speed km/h
    base_time = (distance / base_speed) * 60  # convert hours to minutes

    # Traffic delay
    if traffic.lower() == "high":
        base_time += 15
    elif traffic.lower() == "medium":
        base_time += 8

    # Weather delay
    if weather.lower() == "rainy":
        base_time += 10
    elif weather.lower() == "stormy":
        base_time += 20

    print("Estimated Delivery Time:", round(base_time), "minutes")


# Sample Run
estimate_delivery_time(8, "High", "Rainy")
```

 Problem Statement 3:Movie Theatre Seat Occupancy Analyzer

```python
def analyze_seating(total_seats, booked_seats):
    booked_count = len(booked_seats)
    occupancy = (booked_count / total_seats) * 100

    print("Occupancy:", round(occupancy), "%")

    if occupancy == 100:
        print("Show Status: Housefull")
    elif occupancy >= 75:
        print("Show Status: Almost Full")
    else:
        print("Show Status: Seats Available")

    if occupancy >= 90:
        print("Suggestion: Consider opening additional show")


# Sample Run
total_seats = 200
booked = [1] * 150
analyze_seating(total_seats, booked)
```

 Problem Statement 4:Cloud Server Load Classification System

```python
def classify_server_load(cpu_readings):
    total = 0

    for value in cpu_readings:
        total += value

    average = total / len(cpu_readings)

    print("Average CPU Load:", round(average), "%")

    if average < 50:
        print("Server Status: Normal")
    elif 50 <= average < 80:
        print("Server Status: Warning")
    else:
        print("Server Status: Critical")


# Sample Run
readings = [45, 60, 70, 85, 90]
classify_server_load(readings)
```

 Problem Statement 5:Smart Classroom Resource Usage Monitor

```python
def monitor_resources(resource_usage):
    overused = []
    threshold = 8  # hours threshold

    for resource, hours in resource_usage.items():
        if hours > threshold:
            overused.append(resource)

    if overused:
        print("Overused Resources:", ", ".join(overused))
        print("Energy Alert: Yes")
    else:
        print("Overused Resources: None")
        print("Energy Alert: No")


# Sample Run
usage = {
    "Projector": 6,
    "AC": 9,
    "Lights": 4
}

monitor_resources(usage)
```

Problem Statement 6: Online Event Registration Capacity Controller

```python
def manage_registrations(capacity, registrations):
    if registrations <= capacity:
        confirmed = registrations
        waitlisted = 0
        status = "Open"
    else:
        confirmed = capacity
        waitlisted = registrations - capacity
        status = "Closed"

    print("Confirmed Registrations:", confirmed)
    print("Waitlisted Users:", waitlisted)
    print("Registration Status:", status)


# Sample Run
manage_registrations(100, 105)
```
