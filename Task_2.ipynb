1. Unique Words in a sentence

```python
sentence = input("Enter sentence:")
words = sentence.split()
unique_words = set(words)
print("Unique words count:",len(unique_words))
print("Unique words:",unique_words)
```

2. Highest Salary From Employee Data

```python
employees = eval(input("Enter employee name and salary:"))
highest_name = max(employees, key=employees.get)
highest_salary = employees[highest_name]
print("Highest Salary:",highest_name,"-",highest_salary)
```

3.Find Maximum And Minimum Values

```python
numbers = list(map(int ,input("Enter the list of numbers:").split(",")))
print("List:",numbers)
print("Max:",max(numbers))
print("Min:",min(numbers))
```

4. Count Products Above a Price Threshold

```python
prices = list(map(int,input("Enter the list of numbers:").split(",")))
count = 0
for price in prices:
    if price >1000:
        count += 1
print("Products above:",count)
```

5. Calculate Attendance Percentage

```python
attendance = eval(input("Enter"))
present_count = 0
for day in attendance:
    if day == "P":
        present_count += 1
percentage = (present_count / len(attendance)) *100
print("Attendance Percentage:", percentage)
    
```

6. Remove Duplicate Phone Numbers

```python
phone_numbers = eval(input("Enter the phone numbers:"))
unique_phone_numbers = set(phone_numbers)
print("Unique phone numbers:", unique_phone_numbers)
```

7. Count Character Frequency

```python
string  = input("Enter a string:")
dict = {}
for ch in string:
    dict[ch] = dict.get(ch,0) + 1
print(dict)
```

10. Calculate Average Marks

```python
marks = list(map(int, input("Enter marks :").split(",")))
avg = sum(marks) / len(marks)
print("Average Marks:",avg)
```
