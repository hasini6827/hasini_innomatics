```python
class SmartLight:
    def __init__(self, name):
        self.name = name
        self.state = "OFF"
    def control(self, action):
        action = action.strip().upper()
        if action in ["ON", "OFF"]:
            self.state = action
    def display(self):
        print(f"{self.name} is {self.state}")
light_name = input("Light Name: ")
action = input("Action: ")
light = SmartLight(light_name)
light.control(action)
light.display()
```

```python
class Employee:
    def __init__(self, name, emp_id, dept):
        self.name = name
        self.emp_id = emp_id
        self.dept = dept
    def show_details(self):
        print("Employee ID Card")
        print("Name:", self.name)
        print("ID:", self.emp_id)
        print("Department:", self.dept)
name = input("Employee Name: ")
emp_id = input("Employee ID: ")
dept = input("Department: ")
emp = Employee(name, emp_id, dept)
emp.show_details()
```

```python
class Contact:
    def __init__(self, name, phone):
        self.name = name
        self.phone = phone
    def display_contact(self):
        print("Contact Saved")
        print("Name:", self.name)
        print("Phone:", self.phone)
name = input("Contact Name: ")
phone = input("Phone Number: ")
c = Contact(name, phone)
c.display_contact()
```

```python
class Product:
    def __init__(self, name, price):
        self.name = name
        self.price = price
    def print_tag(self):
        print("Product:", self.name)
        print("Price: " + str(self.price))
name = input("Product Name: ")
price = int(input("Price: "))
p = Product(name, price)
p.print_tag()
```

```python
class Movie:
    def __init__(self, title, rating):
        self.title = title
        self.rating = rating
    def display_movie(self):
        print("Movie:", self.title)
        print("Rating:", self.rating, "/ 5")
title = input("Movie Name: ")
rating = float(input("Rating: "))
m = Movie(title, rating)
m.display_movie()
```

```python
class Delivery:
    def __init__(self, customer, address):
        self.customer = customer
        self.address = address
    def show_delivery(self):
        print("Delivery Details")
        print("Customer:", self.customer)
        print("Address:", self.address)
customer = input("Customer Name: ")
address = input("Address: ")
d = Delivery(customer, address)
d.show_delivery()
```
