**Problem Statement 1: Employee Performance Bonus Eligibility**

```python
employees = {
    "Ravi":92,
    "Anita":88,
    "Kiran":92,
    "Suresh":85
}
#Identify the highest performance score.
highest_performance_score = max(employees.values())
#Handle ties if multiple employees have the same highest score.
highest_scores  = [name for name, score in employees.items() if score == highest_performance_score]
#Display all employees eligible for the top performance bonus.
print("Top Performers Eligible for Bonus:",",".join(highest_scores),f"(Score: {highest_performance_score})")
```

**Problem Statement 2: Search Query Keyword Analysis**

```python
inp = "Buy mobile phone buy phone online"
#Covert the input into lower case
inp = inp.lower()
#Ignore common punctuation
import string
for char in string.punctuation:
    inp = inp.replace(char,"")
words = inp.split()
frequency = {}
for word in words:
    frequency[word] = frequency.get(word, 0)+1
result = {word: count for word, count in frequency.items() if count>1}
print(result)
```

**Problem Statement 3: Sensor Data Validation**

```python
sensor_readings = [3,4,7,8,10,12,5]
valid_sensor_readings = []
#Identify readings that are even numbers (valid readings).
for index, value in enumerate(sensor_readings):
    if value%2 == 0:
        valid_sensor_readings.append((index,value))
print("Valid Sensor Readings (Hour, Value):")
print(valid_sensor_readings)
    
```

**Problem Statement 4: Email Domain Usage Analysis**

```python
emails = [
"ravi@gmail.com",
"anita@yahoo.com",
"kiran@gmail.com",
"suresh@gmail.com",
"meena@yahoo.com"
]
domain_count = {}
for email in emails:
    domain = email.split("@")[1]
    domain_count[domain] = domain_count.get(domain, 0) + 1
total = len(emails)
for domain, count in domain_count.items():
    percentage = (count / total) * 100
    print(f"{domain}:{percentage:.0f}%")
```

**Problem Statement 5: Sales Spike Detection**

```python
sales = [1200, 1500, 900, 2200, 1400, 3000]
avg = sum(sales) / len(sales)
limit = avg *1.3
for i in range(len(sales)):
    if sales[i] > limit:
        print(f"Day{i + 1}: {sales[i]}")
```

**Problem Statement 6: Duplicate User ID Detection**

```python
user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]
count = {}
for user in user_ids:
    count[user] = count.get(user, 0) + 1
for user, times in count.items():
    if times>1:
        print(f"{user} -> {times} times")
```
